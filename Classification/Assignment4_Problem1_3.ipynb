```python
#create a dummy variable called INJURY that takes the value “yes” if MAX_SEV_IR = 1 or 2, and otherwise “no.”
data["INJURY"] = (data['MAX_SEV_IR'] > 0).astype(int)
# dropping off MAX_SEV_IR column
data = data.drop(list(["MAX_SEV_IR"]), axis = 1)
data.loc[data["INJURY"] == 1, "INJURY"] = "YES"
data.loc[data["INJURY"] == 0, "INJURY"] = "NO"
```

```python
data.info()
```

```python
predictors = ['HOUR_I_R', 'ALCHL_I', 'ALIGN_I', 'STRATUM_R', 'WRK_ZONE', 'WKDY_I_R',
              'INT_HWY', 'LGTCON_I_R', 'MANCOL_I_R', 'PED_ACC_R', 'RELJCT_I_R',
              'REL_RWY_R', 'PROFIL_I_R', 'SPD_LIM', 'SUR_COND', 'TRAF_CON_R',
              'TRAF_WAY', 'VEH_INVL', 'WEATHER_R', 'INJURY_CRASH', 'NO_INJ_I',
              'PRPTYDMG_CRASH', 'FATALITIES']

# All variables are categorical

X = data[predictors]
Y = data['INJURY']
```

```python
# Partition the data into training (60%) and validation (40%).
X_train, X_valid, y_train, y_valid = train_test_split(X, Y, test_size=0.40, random_state=1)
```

```python
# Run a naive Bayes classifier on the complete training set with the relevant predictors
injurys_nb = MultinomialNB(alpha=0.01)
injurys_nb.fit(X_train, y_train)
```

```python
# Return Probability estimates
predProb_train = injurys_nb.predict_proba(X_train)
predProb_valid = injurys_nb.predict_proba(X_valid)
```

```python
# Predict classification
y_valid_pred = injurys_nb.predict(X_valid)
y_train_pred = injurys_nb.predict(X_train)
```

```python
df_valid=pd.DataFrame()
df_valid['class']=y_valid
df_valid['proba']=predProb_valid[:,1]
df_valid['y_true']=[1 if p == 'YES' else 0 for p in df_valid['class']]
df_valid['y_pred']=[1 if p > 0.7 else 0 for p in df_valid.proba]
```

```python
df_valid.head()
```

```python
cm=confusion_matrix(y_valid,y_valid_pred)
cm
true_positive=cm[1][1]
print("true positive: "+str(true_positive))

false_positive=cm[0][1]
print('false positive: '+str(false_positive))

false_negative=cm[1][0]
print('false negative: '+str(false_negative))

true_negative=cm[0][0]
print('true negative: '+str(true_negative))

print(accuracy_score(df_valid['y_true'],df_valid['y_pred']))

print((true_positive+true_negative)/df_valid['class'].count())

print(precision_score(df_valid['y_true'],df_valid['y_pred']))

print(true_positive / (true_positive + false_positive))

#recall / sensitivity
print(recall_score(df_valid['y_true'],df_valid['y_pred']))

print(true_positive / (true_positive + false_negative))
```

```python
fpr, tpr, _ = roc_curve(df_valid.y_true, df_valid.proba)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=[5, 5])
lw = 2
plt.plot(fpr, tpr, color='darkorange',
         lw=lw, label='ROC curve (area = %0.4f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (1 - Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')
plt.legend(loc="lower right")

plt.show()
```

```python
precision_recall_curve(df_valid.y_true, df_valid.proba)

precision, recall, _ = precision_recall_curve(df_valid.y_true, df_valid.proba)

plt.figure(figsize=[5, 5])
lw = 2
plt.plot(precision, recall, color='darkorange',
         lw=lw, label='precision recall curve)')
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.legend(loc="lower right")

plt.show()
```

```python
data_cars=pd.concat([data_cars, pd.get_dummies(data_cars['Fuel_Type'], prefix_sep='_',)],axis=1)


data_cars = data_cars.drop(list(["Id",'Fuel_Type', 'Model', 'Color']), axis = 1)

scaler = preprocessing.MinMaxScaler()
scaled_data = pd.DataFrame(scaler.fit_transform(data_cars), index=data_cars.index, columns=data_cars.columns)
```

```python

predictors = ["Age_08_04", "KM", "Diesel", "Petrol", "HP", "Automatic", "Doors", "Quarterly_Tax", "Mfr_Guarantee", "Guarantee_Period", "Airco", "Automatic_airco", "CD_Player", "Powered_Windows", "Sport_Model", "Tow_Bar"]
outcome = "Price"

X = scaled_data[predictors]
y = data_cars[outcome]
train_X, valid_X, train_y, valid_y = train_test_split(X, y, test_size=0.4, random_state=1)
```

```python
mlp_model = MLPRegressor(hidden_layer_sizes=(1,), activation='relu', solver='lbfgs')

param_grid = {}
gridSearch = GridSearchCV(mlp_model, param_grid, cv=5, n_jobs=-1)
gridSearch.fit(train_X, train_y)

mlp_model = gridSearch.best_estimator_

print(mlp_model)
print("RMSE of Validation: "+str((mean_squared_error(valid_y, mlp_model.predict(valid_X))**.5)))
print("RMSE of Training: "+str((mean_squared_error(train_y, mlp_model.predict(train_X))**.5)))
```

```python
mlp_model = MLPRegressor(hidden_layer_sizes=(2,), activation='relu', solver='lbfgs')

param_grid = {}
gridSearch = GridSearchCV(mlp_model, param_grid, cv=5, n_jobs=-1)
gridSearch.fit(train_X, train_y)

mlp_model = gridSearch.best_estimator_

print(mlp_model)
print("RMSE of Validation: "+str((mean_squared_error(valid_y, mlp_model.predict(valid_X))**.5)))
print("RMSE of Training: "+str((mean_squared_error(train_y, mlp_model.predict(train_X))**.5)))
```
